# Ders 1 – Array (Dizi) ve Dinamik Array

## 1. Array (Statik Dizi) – Teori Özeti

**Ne işe yarar?**
- Aynı tipte verileri **yan yana**, **sabit boyutlu** bir blokta tutar.
- Örnek: 100 öğrencinin notları, 10 sensörün ölçümleri, vs.

**Özellikler:**
- Bellekte **ardışık (contiguous)** saklanır.
- Her elemana **indis** (index) ile O(1) zamanda erişirsin: `a[i]`.
- Boyutu **derleme zamanında** sabittir (classic C tarzı array).

**Avantajlar:**
- Çok hızlı erişim (index ile direkt gitme).
- Cache dostu, performanslı.
- Basit ve düşük seviye; algoritma temeli için ideal.

**Dezavantajlar:**
- Boyut sabit; sonradan büyütüp küçültmek yok.
- Ortaya eleman eklemek/silmek pahalı (sağa sola kaydırma O(n)).


## 2. Dinamik Array – Teori Özeti

**Ne işe yarar?**
- Boyutu başta sabit değil, **ihtiyaca göre büyütülen/küçültülen** dizi.
- C++'ta `std::vector` bunun hazır hali, biz eğitim için basit bir versiyonunu yazacağız.

**Nasıl çalışır (basit mantık)?**
- İçeride yine **normal bir dizi** var (`T* data`).
- Ekledikçe doluyor, kapasite biterse:
  - Daha büyük bir blok alıyoruz (`new`).
  - Eski elemanları kopyalıyoruz.
  - Eski bloğu `delete[]` ile siliyoruz.

**Avantajlar:**
- Eleman ekleyebiliyorsun (`push_back`) → amortized O(1).
- Index ile yine O(1) erişim.

**Dezavantajlar:**
- Büyütme sırasında kopyalama maliyeti (bazı eklemeler O(n)).
- İçerideki kapasite yönetimi ile ilgilenmen gerekiyor.


## 3. Bu derste göreceklerimiz

1. Statik dizi ile basit örnekler:
   - Elemanlara erişme
   - Toplam alma, maksimum bulma, ters çevirme

2. Kendi basit `DynamicArray<T>` sınıfımız:
   - `push_back`, `pop_back`
   - `operator[]`
   - `size()`, `capacity()`

3. Zaman karmaşıklığı kabaca:
   - Index erişimi: O(1)
   - Baştan/ortadan ekleme-silme: O(n)
   - `push_back` (dinamik dizi): amortized O(1)
