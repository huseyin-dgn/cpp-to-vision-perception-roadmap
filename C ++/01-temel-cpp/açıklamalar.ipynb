# 1-) ⭐ #include NEDİR?

>#include → bir kütüphaneyi (başlık dosyasını) programına eklemek demektir.

* C++’ta her özellik otomatik gelmez.
* Bir şeyi kullanmak istiyorsan önce onun başlık dosyasını (header file) eklemek zorundasın.

Örnek:

* std::cout → iostream içinde

* std::string → string içinde

* std::vector → vector içinde

* std::thread → thread içinde

* std::map → map içinde

* std::cout << → iostream + ostream içinde

Yani include şu anlama gelir:

>“Kardeşim, ben bu dosyada şu içerikleri kullanacağım, onları da koduma ekle.”

### ⭐ C++’ta include zorunlu neden?

* C++ derleyicisi çok katı bir dil.
* Her sembolün (cout, string, vector…) nereden geldiğini bilmek zorunda.

Buna şöyle bak:

> ⚙️ C++ derleyicisi:

**"cout mu? Bu ne? Nerede tanımladın? Hangi dosyada? Bana söyle yoksa çalıştırmam."**

### ⭐ Peki neden HER hücrede include yazıyoruz?

* Çünkü C++ hücre mantığında her hücre ayrı bir derlenebilir program gibi davranır.

* Yani hücreler Python gibi ortak ortam paylaşmaz.
* Bu iyi bir şey, çünkü C++ çok daha sıkı bir derleme ister.

* Her hücreyi ayrı bir "küçük main.cpp" gibi düşün.

Bu yüzden:

>#include <iostream> yazmazsan cout tanınmaz

>#include <string> yazmazsan string tanınmaz

# 2-) Cout vs Cin 

## ✔️ std::cout NEDİR?

* cout → Console OUT
>Yani konsola veri gönderen bir çıktı akışı (output stream).

>C++’ta ekrana yazmak dediğimiz şey aslında “bir akışa veri göndermek”.

#### Özetle:

* cout = bilgisayara: “Şunları ekrana bas komutu gönderdiğin musluk

><< = bu musluğa veri akıttığın yön oku


## ✔️ std::cin NEDİR?

* cin → Console IN
> Yani klavyeden veri alan bir giriş akışı (input stream).

#### Kısaca:

* cin = bilgisayara: “Kullanıcıdan veri al” dediğin musluk

> ">>" = kullanıcının verdiği veriyi musluktan içeri çekme oku

### ✔️ O zaman özet:

| Kullanım          | Anlamı                               | Ok Yönü               |
| ----------------- | ------------------------------------ | --------------------- |
| `cout << veri`    | Veriyi ekrana gönder                 | Veri sola → sağa akar |
| `cin >> değişken` | Kullanıcıdan veri alıp değişkene koy | Veri sağ → sola akar  |

## Proje sonunda elimizde bulunan ve Bölüm -1 'de bulunan bütün işlemlerin olduğu kod parçası ektedir ;

```cpp
// %% DERS 1 — Hücre 10
#include <iostream>
#include <string>

int main() {
    std::string plaka = "BOS";
    int hiz = 0;
    double motor = 0.0;

    int secim = 0;

    do {
        std::cout << "\n=== TELEMETRI MENUSU ===\n";
        std::cout << "1) Arac bilgisi gir\n";
        std::cout << "2) Telemetri ozetini goster\n";
        std::cout << "3) Cikis\n";
        std::cout << "Secim: ";
        std::cin >> secim;

        if (secim == 1) {
            std::cout << "Plaka: ";
            std::cin >> plaka;

            std::cout << "Hiz (km/s): ";
            std::cin >> hiz;

            std::cout << "Motor sicakligi (C): ";
            std::cin >> motor;
        }
        else if (secim == 2) {

            if (plaka == "BOS") {
                std::cout << "Henuz arac bilgisi girilmedi.\n";
            } else {
                std::cout << "\n--- TELEMETRI OZETI ---\n";
                std::cout << "Plaka: " << plaka << std::endl;
                std::cout << "Hiz: " << hiz << " km/s" << std::endl;
                std::cout << "Motor: " << motor << " C" << std::endl;

                if (hiz > 130)
                    std::cout << "UYARI: YUKSEK HIZ!\n";
                else if (hiz > 90)
                    std::cout << "Uyari: Hiz limitini astiniz.\n";

                if (motor > 100)
                    std::cout << "UYARI: Motor SICAK!\n";
                else if (motor < 60)
                    std::cout << "Bilgi: Motor soguk.\n";
            }
        }
        else if (secim == 3) {
            std::cout << "Cikis yapiliyor...\n";
        }
        else {
            std::cout << "Gecersiz secim!\n";
        }

    } while (secim != 3);

    return 0;
}
